### Import matplotlib module say pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Define arbitrary data points. It will draw only two points.

```python
xpoints = np.array([0,6])
ypoints = np.array([0,250])
```

### Plot the data with line

```python
plt.plot(xpoints,ypoints)
plt.show()
```

### Plot the data without line or markers

```python
plt.plot(xpoints,ypoints,'o')
plt.show()
```
