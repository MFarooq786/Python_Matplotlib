```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
```

```python
plt.plot(x, y)
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')
plt.title('Sports watch Data')

plt.grid()
plt.show()
```

### Specify Which Grid Lines to Display 
You can use axis parameter in the grid() function to specify which grid lines to display. 

```python
plt.plot(x, y)
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')
plt.title('Sports watch Data')

plt.grid(axis='x')
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')
plt.title('Sports watch Data')

plt.grid(axis='y')
plt.show()
```

### Set Line Properties for the Grid 
You can also set the line properties of the grid like: grid(color='color', linestyle='linstyle',linewidth=number)

```python
plt.plot(x, y)
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')
plt.title('Sports watch Data')

plt.grid(color='r',linestyle='--', linewidth=0.5)
plt.show()
```
