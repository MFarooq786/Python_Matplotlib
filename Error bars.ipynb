```python
import matplotlib.pyplot as plt

# Data points
y = [55.9453 , 62.2147, 87.8038, 94.2884]

# Errors for each data point
yerr = [0.588646, 0.328079, 0, 0.967595]

# X values (assuming sequential for demonstration)
x = range(1, len(y)+1)

# Create the error bar plot
plt.errorbar(x, y, yerr=yerr, fmt='o', ecolor='orangered', capsize=5, linestyle='None', marker='s', markersize=5, color='skyblue')

# Add labels and title for clarity
plt.xlabel('Data Point')
plt.ylabel('Value')
plt.title('Error Bar Plot of Data Points')

# Show the plot
plt.show()
```
