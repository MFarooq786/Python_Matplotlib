### Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Linestyle
You can use keyword argument linestyle or ls to change the style of the line.

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, ls='-.')
plt.show()
```

### Line Color
You can use keyword argument color or c to set the line color. 

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, c='r')
plt.show()
```

### Line Width
You can use keyword argument linewidth or lw to set the line width. 

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, lw=10)
plt.show()
```

### Multiple Lines
You can plot as many lines as you like by simply adding more plt.plot() functions:

```python
ypoints1 = np.array([3, 8, 1, 10])
ypoints2 = np.array([6, 2, 7, 11])
plt.plot(ypoints1)
plt.plot(ypoints2)

plt.show()
```
