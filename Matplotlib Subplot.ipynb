### Display multiple plots
With the subplot() function you can draw multiple plots in one figure:

### Draw 2 plots:

```python
# Plot 1
y = np.array([3, 8, 1, 10])
x = np.array([0, 1, 2, 3])

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('Sales')
# plot 2
y = np.array([10, 20, 30, 40])
x = np.array([0, 1, 2, 3])

plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title('Income')
plt.show()
```

### Super Title
You can add a title to the entire figure with the supertitle() function:

```python
# Plot 1
y = np.array([3, 8, 1, 10])
x = np.array([0, 1, 2, 3])

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('Sales')
# plot 2
y = np.array([10, 20, 30, 40])
x = np.array([0, 1, 2, 3])

plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title('Income')

plt.suptitle('MY Shop')
plt.show()
```
