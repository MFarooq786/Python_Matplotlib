### Creating Bars
You can use bar() function to draw bar graphs:

```python
# Draw 4 bars:
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A","B","c","D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
x = ['apple','bananan']
y= [400, 350]

plt.bar(x,y)
```

### Horizontal Bars
If you want to draw horizontal bars use barh()

```python
x = np.array(["A","B","c","D"])
y = np.array([3, 8, 1, 10])

plt.barh(x,y)
plt.show()
```

```python
x = np.array(["A","B","c","D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y, width=0.1, color='red')
plt.show()
```

```python
x = np.array(["A","B","c","D"])
y = np.array([3, 8, 1, 10])

plt.barh(x,y, height=0.1)
plt.show()
```
