```python
import matplotlib.pyplot as plt
import numpy as np
```

You can use the keyword argument marker to emphasize each point with a specified marker
https://www.w3schools.com/python/matplotlib_markers.asp

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker='o')
plt.show()
```

### Format String
You can also use the shortcut string notation parameter to specify the marker. This parameter is also called fmt which is marker|line|color

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, 'o-.r')
plt.show()
```

### Marker size
You can use the keyword argument markersize or ms to set the size of the marker. 

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker='o',ms=5)
plt.show()
```

### Marker Edge Color
You can use the keyword markeredgecolor or mec to set the color of the edge of the markers: 

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker='o', ms=10, mec='r')
plt.show()
```

### Marker Face Color
You can use keyword markerfacecolor or mfc to set the marker face color. 

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker='o', ms='15', mec='b',mfc='r')
plt.show()
```
